'-'를 기준으로 문자열을 나눠 저장했더니                     
1번 주제-1번 주제에 대한 설명, 2번 주제-2번 주제에 대한 설명, ... 형태가                
1번 주제 | 1번 주제에 대한 설명, 2번 주제 | 2번 주제에 대한 설명, 3번 주제 | ... 형태로 저장되었다.             
→ [(n-1)번 주제에 대한 설명, n번 주제] 형태에서 주제에 대한 설명 부분만 남겨야 함.          
→ rfind 함수로 오른쪽에서 부터 ','를 찾고 ','가 있다면 , 전까지의 문자열을 저장하고, ',' 없다면 전체 문자열을 저장.

### 자연어 처리 (텍스트 전처리)

토큰화(Tokenization): 문장을 단어나 구(phrase)로 분할하여 토큰(token)으로 나눕니다.                 
불용어 제거(Stopword Removal): 자주 등장하지만 분석에 별 도움이 되지 않는 단어들을 제거합니다.                  
단어의 정규화(Normalization): 단어들을 어근 형태로 변환하거나, 대소문자를 통일하는 등의 작업을 수행합니다.                  
품사 태깅(Part-of-Speech Tagging): 각 단어의 품사를 식별합니다.                 
명사구 추출(Noun Phrase Extraction): 문장에서 명사구를 식별하여 추출합니다.                 
개체명 인식(Named Entity Recognition, NER): 이름, 날짜, 위치, 조직 등과 같은 개체명을 인식합니다.                   
감성 분석(Sentiment Analysis): 문장의 감정이나 의견을 분석하여 긍정적인지 부정적인지 판별합니다.                    
토픽 모델링(Topic Modeling): 문서 집합에서 주제를 추출합니다.                   
문서 벡터화(Document Vectorization): 텍스트 데이터를 수치형 벡터로 변환하여 기계학습 알고리즘에 입력할 수 있도록 합니다.

### About TfidfTransformer vs TfidfVectorizer
So we have got the same result, looks good. One important thing I would like to add here is both Tfidfvectorizer() and Tfidftransformer() perform the same calculation only difference is Tfidftransformer acts on sparse matrix and Tfidfvectorizer acts on raw text data.
- Tfidfvectorizer는 원시 텍스트 데이터를 입력받아 직접 tf-idf 벡터를 생성하고 별도의 토큰화나 빈도수 계산 과정이 불필요함.
- Tfidftransformer는 이미 토큰화되고 단어 빈도수가 계산된 데이터에 대해 tf-idf 가중치 적용할 때 사용함.

그러나 TfidfVectorizer는 텍스트를 공백이나 구두점을 기준으로 나눠 토큰화하고 각 토큰의 빈도를 기반으로 tfidf 값을 계산하는 것이기 때문에 한글의 경우에는 제대로 처리되지 않을 수 있음.

한글어 전처리의 경우 형태소 분석-품사태깅 및 필터링-사용자 정의 토크나이저(tokenizer 파라미터를 통해 지정해서 사용)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# 불용어 목록 생성
noun_stop_words = ['제주', '제주도', '곳', '것', '가게', '소품', '등']
morph_stop_words = [',','.', '은', '는', '이', '가', '을', '를', '의', '로', '에서', '수', '에', '으로']

# 토크나이저 생성
okt = Okt()

# 텍스트 형태소 분리 및 명사 추출 (비교를 위해 일단 둘 다 써본다)
def morphTokenizer(text) : 
    text_morphs = okt.morphs(text)
    # 불용어 제거
    filtered_morphs = [morphs for morphs in text_morphs if not morphs in morph_stop_words]
    return ' '.join(filtered_morphs)

def nounTokenizer(text) : 
    # 텍스트 형태소 분리 및 명사 추출 (비교를 위해 일단 둘 다 써본다)
    text_nouns = okt.nouns(text)
    # 불용어 제거
    filtered_nouns = [nouns for nouns in text_nouns if not nouns in noun_stop_words]
    return ' '.join(filtered_nouns)
```

```python
# Tf-Idf 초기화
morph_vectorizer = TfidfVectorizer(tokenizer=morphTokenizer, stop_words=morph_stop_words)
noun_vectorizer = TfidfVectorizer(tokenizer=nounTokenizer, stop_words=noun_stop_words)
```

```python
# 입력한 문장과 유사한 관광지 추출
def similar_spots(input_text) : 
    # 사용자 입력 텍스트 벡터화
    vectorized_input_text = noun_vectorizer.transform([input_text])

    # 사용자 입력 텍스트와 관광지 상세 설명 간의 유사도 계산
    cosine_sim = cosine_similarity(vectorized_input_text, nounVec)
    print('cosine_sim : ', cosine_sim)

    # 유사도 높은 20개 인덱스 찾기
    sim_indexes = np.argsort(cosine_sim[0])[-20:][::-1]         # [::-1] : 내림차순으로 정렬
    result_spots = jejuSpots.iloc[sim_indexes]
    print(sim_indexes)
    result_spots['similarity'] = cosine_sim[0][sim_indexes]

    # 해당 인덱스의 데이터프레임 반환
    return result_spots
```

머신 러닝으로 적용 : 사용자의 입력과 데이터셋 내의 항목들을 클러스터링하여 비슷한 그룹으로 묶고, 각 그룹 내에서 유사도를 계산하여 추천하는 방식

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# 가정: 가지고 있는 리뷰 데이터와 지역 정보 데이터
review_data = ["여기 정말 좋다", "다시 가고 싶은 곳", "최고의 여행지", "맛있는 식당", "아름다운 경치", "편안한 휴식을 즐길 수 있는 곳"]
jejuSpots = pd.DataFrame({
    '스팟': ['장소A', '장소B', '장소C', '장소D', '장소E', '장소F'],
    '설명': review_data
})

# 벡터라이저 초기화 및 학습
noun_vectorizer = TfidfVectorizer(tokenizer=nounTokenizer, stop_words=noun_stop_words)
nounVec = noun_vectorizer.fit_transform(review_data)

def similar_spots(input_text):
    # 사용자 입력 텍스트 벡터화
    vectorized_input_text = noun_vectorizer.transform([input_text])

    # 사용자 입력 텍스트와 리뷰 데이터 간의 유사도 계산
    cosine_sim = cosine_similarity(vectorized_input_text, nounVec)
    print('cosine_sim : ', cosine_sim)

    # 유사도 높은 20개 인덱스 찾기
    sim_indexes = np.argsort(cosine_sim[0])[-20:][::-1]
    result_spots = jejuSpots.iloc[sim_indexes]
    print(sim_indexes)
    
    # 해당 인덱스의 데이터프레임 반환
    return result_spots

inputText = "데이트 하기 좋은 곳"
print(similar_spots(input_text=inputText))
```
