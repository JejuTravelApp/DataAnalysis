### 컬럼별 전처리 계획
- title 컬럼의 NA 데이터 → 삭제 처리          
- stars 컬럼            
    NA 데이터 → NA 대신 '정보 없음' 으로 바꾸기              
    notNA → int or double type으로 바꾸기            
- mainTag & subTag → 두 컬럼 합치고 중복된 단어 제거하기                 
- text 컬럼의 NA → 6개 데이터 중 3개 : 숙소, 1개 : 쇼핑, 1개 : 식당 1개 : 관광지(공원).         
    공백 처리       
    공백일 경우 BodyRightText 컬럼의 소개 문장으로 모델 만들기          
- BodyRightText 컬럼 : 소개 부분만 잘라내서 introduction 컬럼 재생성하기 / NA -> 공백 처리

### 컬럼별 전처리

```python
visitJejuSpots = df.copy()
visitJejuSpots.head()
```

#### title 컬럼
NA 데이터 삭제

```python
len(visitJejuSpots)
```

```python
visitJejuSpots[visitJejuSpots['title'].isna()]
```

```python
visitJejuSpots.dropna(subset=['title'], inplace=True)
```

```python
len(visitJejuSpots)
```

### stars 컬럼
- NA → ' ' (공백) 
- notNA → 평가 점수만 추출(평가점수 string 타입으로 저장, 후에 순서 비교할때 int로 바꿔 비교하기)

```python
visitJejuSpots.count()
```

```python
# notNA 데이터 int type으로 값 뽑아내기
visitJejuSpots['stars'].unique()
```

```python
visitJejuSpots['stars'].unique
```

```python
visitJejuSpots['stars'].str[-3]
```

```python
visitJejuSpots['stars'] = visitJejuSpots['stars'].str[-3]
visitJejuSpots[visitJejuSpots['stars'].notna()]
```

```python
# NA 데이터 공백으로 채우기
visitJejuSpots['stars'] = visitJejuSpots['stars'].fillna('')
visitJejuSpots
```

#### mainTag & subTag 
합치고 중복된 단어 제거하기

```python
visitJejuSpots.count()
```

```python
print(len(visitJejuSpots[visitJejuSpots['mainTag'].isna()]))
print(len(visitJejuSpots[visitJejuSpots['subTag'].isna()]))
```

```python
# 두 컬럼의 데이터 수가 다르면 합쳐지지 않으므로 subTag 컬럼의 na를 공백으로 채워줌
visitJejuSpots['subTag'].fillna('', inplace=True)
print(len(visitJejuSpots[visitJejuSpots['subTag'].isna()]))
```

```python
# # test
# data_test = {'A': ['#가족 #부모', '#연인 #과자 #조용히'],
#          'B': ['#가족 #연인', '#부모 #조용히']}
# dfa = pd.DataFrame(data_test)

# # 각 행에 대해 중복을 제외하고 합치기
# dfa['C'] = dfa.apply(lambda row: ' '.join(set(row['A'].split() + row['B'].split())), axis=1)
# # 결과 출력
# dfa
```

```python
# 두 컬럼의 데이터를 합치고 중복된 단어 제거하기
visitJejuSpots['tag'] = visitJejuSpots.apply(lambda row: ' '.join(set(row['mainTag'].split() + row['subTag'].split())), axis=1)
visitJejuSpots
```

### text 컬럼
공백처리

```python
visitJejuSpots.count()
```

```python
visitJejuSpots['text'].fillna('', inplace=True)
visitJejuSpots.count()
```

### BodyRightText 컬럼
NA → 공백처리
소개 부분만 잘라서 introduction 컬럼 생성

```python
visitJejuSpots[visitJejuSpots['BodyRightText'].isna()]
```

```python
visitJejuSpots['BodyRightText'].fillna('', inplace=True)
visitJejuSpots.count()
```

```python
visitJejuSpots['BodyRightText']
```

```python
# '-'를 기준으로 문자열 나누기
splited_by_bar = visitJejuSpots['BodyRightText'].str.split('-')
print(splited_by_bar)
print("한 줄 : ",splited_by_bar[13])
print("한 줄1 : ",splited_by_bar[137])
print("한 줄2 : ",splited_by_bar[725])
print("한 줄3 : ",splited_by_bar[797][1])
print("만들고자 하는 부분 : ", splited_by_bar[3][1])
```

```python
introductions = []
i = 0
for sentences in splited_by_bar:
    if i == 137 : 
        # print(i, '번째 : 없음')           # 왜 안돼??
        introductions.append('')
    elif i == 725 : 
    # elif sentences == '없음' : 
        # print(i, '번째 : 없음')
        introductions.append('')
    else : 
        # print(sentences[1])
        commaIndex = sentences[1].rfind(',')        # rfind(',') : 문자열에서 오른쪽부터 쉼표(,)를 찾고, 쉼표(,)가 발견되지 않으면 -1을 반환
        if commaIndex != -1 :            
            # print(i, "번째 : ",sentences[1])
            introductions.append(sentences[1][:commaIndex])
        else : 
            # print(i, "번째 : ",sentences[1])
            introductions.append(sentences[1])
        # test_intro.append(sentences[1].split(',')[:-1])           # 상세 정보 이용시간 등 없는 데이터들은 값이 삭제됨.
    i += 1
#     introduction.append(sentences[1])
```
