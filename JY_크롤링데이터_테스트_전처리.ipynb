#### 3-3. SOYNLP의 응집확률(cohension probability)
응집 확률 : 내부 문자열(substring)이 얼마나 응집하여 자주 등장하는지를 판단하는 척도
- 문자열을 문자 단위로 분리하여 내부 문자열을 만드는 과정에서 왼쪽부터 순서대로 문자를 추가하면서 각 문자열이 주어졌을 때 그 다음 문자가 나올 확률을 계산하여 누적곱을 한 값
- 이 값이 높을수록 전체 코퍼스에서 이 문자열 시퀀스는 하나의 단어로 등장할 가능성이 높습니다.
- 응집도의 값을 통해 하나의 단어로 판단하기에 적합한지 판단 가능.

```python
# word_score_table["예시단어"].cohesion_forward
```

#### 3-4. SOYNLP의 브랜칭 엔트로피(branching entropy)
주어진 문자열에서 얼마나 다음 문자가 등장할 수 있는지를 판단하는 척도
- 확률 분포의 엔트로피값을 사용
- 브랜칭 엔트로피를 주어진 문자 시퀀스에서 다음 문자 예측을 위해 헷갈리는 정도로 비유한다면 브랜칭 엔트로피의 값은 하나의 완성된 단어에 가까워질수록 문맥으로 인해 점점 정확히 예측할 수 있게 되면서 점점 줄어드는 양상을 보입니다.
- 문자 시퀀스 다음에는 조사나 다른 단어와 같은 다양한 경우가 있을 수 있기 때문에 하나의 단어가 끝나면 그 경계 부분부터 다시 브랜칭 엔트로피 값이 증가함

#### 3-5. SOYNLP의 L tokenizer
한국어의 경우 공백(띄어쓰기)으로 분리된 하나의 문자열은 'L 토큰 + R 토큰' 구조인 경우가 많다. 왼쪽에 오는 L 토큰은 체언(명사, 대명사)이나 동사, 형용사 등이고 오른쪽에 오는 R 토큰은 조사, 동사, 형용사 등이다.
- L 토크나이저는 L 토큰 + R 토큰으로 나누되, 분리 기준을 점수가 가장 높은 L 토큰을 찾아내는 원리

```python
token_ko = t.nouns(review[0])
token_ko
```

```python
# stop words
stop_words = ['제', '나']
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
token_ko
```

```python
ko = nltk.Text(token_ko)
ko.vocab().most_common()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 상위 50개의 빈도수 차트 만들기
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```
