```python
# alltag 컬럼의 데이터 개수와 tag 컬럼의 데이터 개수가 다름. -> alltag 컬럼의 null 값을 공백으로 만든 뒤 합치기
df['alltag'].fillna('', inplace=True)
# df['tags'] = [', '.join([alltag, tag]) for alltag, tag in zip(df['alltag'], df['tag'])]
df['tags'] = df.apply(lambda row: ', '.join(set(', '.join([row['alltag'], row['tag']]).split(', '))), axis=1)
df.head()
```

### 컬럼명 수정

```python
# 컬럼명 수정하기
jeju.rename({'contentscd.label': 'category', 'tags': 'tag', 'repPhoto.photoid.imgpath': 'imgpath'}, axis=1, inplace=True)
jeju.head()
```

```python
jeju.count()
```

주소와 관련된 컬럼 (address, roadaddress, latitude, longitude) 중 하나라도 null 일 경우의 데이터 확인

```python
jeju_null_addr = jeju[(jeju['address'].isnull()) | (jeju['roadaddress'].isnull()) | (jeju['latitude'].isnull()) | (jeju['longitude'].isnull())]
jeju_null_addr
```

관광지 추천을 위한 분석이기 때문에 분석에 포함시킬지 여부를 직접 검색해서 확인
