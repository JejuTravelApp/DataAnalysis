### 컬럼별 전처리 계획

분석에 주로 쓰이는 컬럼이 review 컬럼이기 때문에 review 컬럼에서 특수문자 및 공백 제거 후 NaN인 데이터의 경우 삭제 처리

```python
df.info()
```

```python
# 특수문자 제거 전
# df['review'][7502]
df['review'][0]
```

```python
df
```

```python
# review 컬럼의 특수문자 제거
import re
df['review'] = df['review'].apply(lambda x: re.sub(r'\\n', '', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'\n', ' ', x))
# df['review'] = df['review'].apply(lambda x: re.sub(r'[^a-zA-Z가-힣0-9\s]', '', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'[^a-zA-Z가-힣0-9%]', ' ', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'\\', '', x))
df.head()
```

```python
# 특수문자 제거 확인
# df.iloc[7502:7505]
df['review'][0]
```

```python
df
```

```python
df.iloc[7502:7505]
```

```python
# 특수문자 제거 후 공백을 NaN으로 변경후 NaN인 리뷰 제거
import numpy as np
df['review'] = df['review'].replace(r'^\s*$', np.nan, regex=True)
df = df.dropna(subset=['review'])
df = df.reset_index(drop=True)
```

```python
# 데이터 프레임 확인
df.info()
```
