### 컬럼별 전처리 계획

분석에 주로 쓰이는 컬럼이 review 컬럼이기 때문에 review 컬럼에서 특수문자 및 공백 제거 후 NaN인 데이터의 경우 삭제 처리

```python
df.info()
```

```python
# 특수문자 제거 전
# df['review'][7502]
df['review'][0]
```

```python
df
```

```python
# review 컬럼의 특수문자 제거
import re
df['review'] = df['review'].apply(lambda x: re.sub(r'\\n', '', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'\n', ' ', x))
# df['review'] = df['review'].apply(lambda x: re.sub(r'[^a-zA-Z가-힣0-9\s]', '', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'[^a-zA-Z가-힣0-9%]', ' ', x))
df['review'] = df['review'].apply(lambda x: re.sub(r'\\', '', x))
df.head()
```

```python
# 특수문자 제거 확인
# df.iloc[7502:7505]
df['review'][0]
```

```python
df
```

```python
df.iloc[7502:7505]
```

```python
# 특수문자 제거 후 공백을 NaN으로 변경후 NaN인 리뷰 제거
import numpy as np
df['review'] = df['review'].replace(r'^\s*$', np.nan, regex=True)
df = df.dropna(subset=['review'])
df = df.reset_index(drop=True)
```

```python
# 데이터 프레임 확인
df.info()
```

### rating의 type이 왜 object인지 확인 후 변경

```python
df['rating'].unique(), df['rating'].max(), df['rating'].min()
```

크롤링으로 가져오지 못한 값을 'x'로 입력하면서 실수와 문자가 혼동되어 object type이 된 듯.          
→ x를 NaN으로 바꾸고 데이터 타입 변형

```python
df['rating'] = df['rating'].replace('x', np.nan)
df['rating'] = df['rating'].astype(float)
df.info()
```

## 토큰화 및 형태소 분리

```python
# 여러 형태소 분석기 비교
from konlpy.tag import Okt
from konlpy.tag import Hannanum    
from konlpy.tag import Komoran
from konlpy.tag import Kkma
import nltk

okt = Okt()
hnn = Hannanum()
kmr = Komoran()
kkm = Kkma()
```

```python
# Okt 분석기
# 형태소 추출
oktMorphs = []
# 품사 태깅
oktPos = []
# 명사 추출
oktNouns = []

# 리뷰 컬럼 형태소 분리하기
for review in jejuRestaurant['review']:
    oktMorphs.append(okt.morphs(review))
    oktPos.append(okt.pos(review))
    oktNouns.append(okt.nouns(review))
```
