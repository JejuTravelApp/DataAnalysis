```python
# 정규 표현식을 사용하여 review의 줄바꿈 문자 제거
def remove_enter_symbol(text):
    return re.sub(r'\n+', ' ', text)

# dfCopy['review'] 컬럼에 대해 각 행별로 함수 적용
review = dfCopy['review'].apply(remove_enter_symbol)
review
```

한글, 영어 제외 모든 문자 제거하기

```python
def remove_non_text(text):
    # 한글과 영어를 제외한 모든 문자 제거
    return re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s,]','', text)

# 리뷰 컬럼에 대해 각 행에 함수 적용
# review = dfCopy['review'].apply(remove_non_text)
review = review.apply(remove_non_text)
review
```

# 일단 모델 만드는거부터 테스트 해보기

```python
# 코사인 유사도 바탕으로 모델 만들기
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(dfCopy['review'])
print("행렬 크기", tfidf_matrix.shape)
```

```python
# 각 식당 벡터에 대해 상호간의 코사인 유사도 계산
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("행렬 크기", cosine_sim.shape)
```

```python
# review_to_index = dict(zip(dfCopy['사업장명'], dfCopy.index))
title_to_index = dict(zip(dfCopy['사업장명'], dfCopy.index))

# 타이틀과 관련된 타이틀의 인덱스?? - 일단 타이틀 입력후 유사한 식당 반환하는 모델 만들어보기 (O)
# idx = review_to_index['커피가 맛있어요']
idx = title_to_index['멜꽃삼']
print(idx)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = title_to_index[title]

    # 모든 사업장의 입력 사업장과의 유사도 계산
    sim_score = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 식당 정렬
    sim_scores = sorted(sim_score, key=lambda x: x[1], reverse=True)

    # 가장 유사한 20개 식당
    sim_scores = sim_scores[1:21]

    # 가장 유사한 20개 식당 인덱스 구성
    restaurant_indices = [idx[0] for idx in sim_scores]

    return dfCopy['사업장명'].iloc[restaurant_indices]
```

```python
recommendations = get_recommendations('멜꽃삼')
recommend_title = list()
for title in recommendations:
    # recommend_title = []
    recommend_title.append(title)
recommend_title
# recommend_title[9]            # 관련있는 10번째 음식점 타이틀
```
