# 크롤링해온 csv파일들을 합치고 전처리하는 과정

```python
import pandas as pd
```

### 크롤링해온 csv가 1000개 단위로 쪼개져있기 때문에 한줄로 합칠 필요가 있음

```python
# 1000개에서 10000개를 한줄로 합치기 위해서 for문으로 한꺼번에 추가해주기
dfs = [] # 데이터프레임을 저장할 리스트

for i in range(1, 11):
    df = pd.read_csv(f"Data/제주도식당_크롤링_{i}000.csv")
    dfs.append(df)

result_df = pd.concat(dfs)
```

```python
result_df.reset_index()
```

# 식당 크롤링 데이터 전처리하기
- 15001 ~ 19000 까지의 데이터는 제외

```python
# 기준 데이터 불러오기
scrapyData = pd.read_csv("Data/제주도식당_전처리.csv")
```

```python
# 15001 ~ 19000 제외하기
# 슬라이싱을 사용하여 특정 범위의 행을 제외하고 다른 행들만 선택
resultData = scrapyData.drop(scrapyData.index[15000:19000])
```

```python
resultData.count(axis=1)
```

```python
# 1000개에서 15000개를 한줄로 합치기 위해서 for문으로 한꺼번에 추가해주기
dfs = [] # 데이터프레임을 저장할 리스트

for i in range(1, 16):
    df = pd.read_csv(f"Data/크롤링 데이터/제주도식당_크롤링_{i}000.csv")
    dfs.append(df)

reviewData = pd.concat(dfs)
```

```python
reviewData.reset_index(inplace=True)
```

```python
reviewData.tail()
```

```python
resultData[14996: 15001]
```

> 15000까지는 전부 일치함

```python
# 19000 ~ 20565 붙여주기
df20 = pd.read_csv(f"Data/크롤링 데이터/제주도식당_크롤링_20000.csv")
```

```python
df20.tail()
```

### finalResultData와 resultData 합치기
- 리뷰 못가져온 장소는 지워버리기

```python
finalResultData.reset_index(inplace=True)
```

```python
resultData.reset_index(inplace=True)
```

```python
concaatResultData = pd.concat([resultData, finalResultData], axis=1)
```

```python
concaatResultData.tail()
```

```python
# 필요 컬럼만 남기기
concaatResultData = concaatResultData[["사업장명", "업종구분대분류", "업종구분소분류", "foodCategory", "소재지전체주소", "도로명전체주소", "lat", "lng", "images", "rating", "visitReviewCount", "blogReviewCount", "review"]]
```

```python
# 결측치 확인
concaatResultData.isna().count()
```

```python
# null값 확인
concaatResultData.isnull().sum()
```

```python
concaatResultData = concaatResultData.fillna("x")
```

```python
# null값 확인
concaatResultData.isnull().sum()
```

```python
# 전부 못가져온거 확인하기
concaatResultData[concaatResultData['review'] == "x"].count(axis=1)
```

> 16656개의 데이터중 8858개의 리뷰만 존재, 7798개는 못가져옴

### 리뷰 na값인 경우에도 지도데이터로는 사용이 가능하니 넣어주고   
### na값을 지운건 학습용 데이터로 사용하기

```python
# Map에 쓸 데이터는 리뷰가 필요 없기 때문에 제거
mapData = concaatResultData.drop(columns=['review'])
```

```python
mapData[mapData['foodCategory'] == "x"].count(axis=1)
```

```python
mapData[13876: 13877]
```

```python
# 지도에 들어갈 모든 음식점 데이터 저장
mapData.to_json("Data/제주도식당_지도Data.json", orient='records', force_ascii=False)
```

```python
# 학습용에 쓸 데이터는 리뷰가 x인건 제외하기
mLData = concaatResultData[concaatResultData['review'] != "x"]
```
